```python
""""Do the necessary imports"""
from bs4 import BeautifulSoup
import requests
import pandas as pd
import os
```

```python
""""Scrape the first page to retrieve the list of all the countries where data is available"""

url = "https://www.gov.uk/foreign-travel-advice"

response=requests.get(url)

soup_1 = BeautifulSoup(response.content, 'html.parser')
```

```python
""""Put the name of all the countries in a list and do modifications so that they fit the url format necessarry for further requests"""

countries_list = [i.text for i in soup_1.find_all(class_="govuk-link countries-list__link")]

countries_list = [i.lower() for i in countries_list]

countries_list = [i.replace(" ", "-") for i in countries_list]

countries_list
```

```python
""""Verify that the country names in the list of countries are correctly spelled and don't return an error page. Correct the mistakes """


response_check = []

for i in countries_list:
    
    url = f"https://www.gov.uk/foreign-travel-advice/{i}"
    
    response=requests.get(url)
    
    #soup = BeautifulSoup(response.content, 'html.parser')
    
    if response.status_code != 200:
        response_check.append(i)
        

corrected_spelling = ["british-antarctic-territory", "bonaire-st-eustatius-saba", "cook-islands-tokelau-and-niue","cote-d-ivoire","curacao", "myanmar", "reunion", "sao-tome-and-principe", "st-helena-ascension-and-tristan-da-cunha", "st-martin-and-st-barthelemy", "st-pierre-and-miquelon"  ]
    
for i in response_check:
    countries_list[countries_list.index(i)]=corrected_spelling[response_check.index(i)]    
    
    
    
   
    
```

```python
"""Scrape the links to the maps and put them in a dict"""

results={}

for i in countries_list:
    
    url = f"https://www.gov.uk/foreign-travel-advice/{i}"
    
    response=requests.get(url)
    
    soup = BeautifulSoup(response.content, 'html.parser')
    
    if soup.find(class_="map") is None:
        results[i.capitalize()]= f"No map of {i.capitalize()} available"
        
    else:
        results[i.capitalize()]=soup.find(class_="map-image").get("src")
```
